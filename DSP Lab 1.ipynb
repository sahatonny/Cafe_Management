```python
#sinusoid
import numpy as np
import matplotlib.pyplot as plt

f=10
t=np.arange(0,1,0.01)
signal = np.sin(2*np.pi*f*t)
plt.plot(signal)
#plt.stem(signal)
```

```python
#unit step
sig = []
sig = [1 for i in range(20)]
plt.stem(sig)
```

```python
#exponential
a=0.9
sig = []
sig = [a**i for i in range(20)]
plt.stem(sig)
```

```python
#unit ramp
sig = []
sig = [i for i in range(20)]
plt.stem(sig)
```

```python
#unit impulse
sig = []
sig =[0 for i in range(20)]
sig[0] = 1
plt.plot(sig)
```

```python
#system response
x = [1,3,4,0,10]
h = [1,2,3]
#y = np.convolve(x,h)
#plt.stem(y)
y = []
for n in range(7):
  sum=0
  for k in range (len(x)):
    if(n-k>=0) and (n-k<len(h)):
      sum+= (x[k]*h[n-k])
  y.append(sum)

plt.stem(y)
```
